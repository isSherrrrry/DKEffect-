```python
file = open('interaxis_2021-02-05_002938_2021-02-09_003038.log', 'r')
lines = file.read().splitlines()
file.close()
```

```python
keep = []
for line in lines:
    if "classification\":\"none" not in line and "eventType\":\"click" in line:
        keep.append(line)
keep
```

```python
short = []
for kept in keep:
    string = kept.partition("whichCondition")[0] 
    short.append(string)
short
```

```python
dict = {}
uid_lst = []
for item in short:
    text = item.partition("classification\":\"")[2]
    classification = text.partition("\",\"userId")[0]
#     print(user_classification)
    uid = text.partition("\",\"userId\":\"")[2][0:-3]
#     print(uid)
    name = item.partition("Name\":\"")[2].partition("\",\"")[0]
#     print(name)
    if uid not in uid_lst:
        uid_lst.append(uid)
    if uid in ["1612648318306"]:
        if uid not in dict:
            dict[uid] = [[name, classification]]
        else:
            dict[uid].append([name, classification])
dict
```

```python
# loading data
import os
import pandas as pd 
dog_data = pd.read_csv("/Users/emily/Desktop/categorization_experiment-master/data/dog_indiv_50_removed.csv") 
ball_data = pd.read_csv("/Users/emily/Desktop/categorization_experiment-master/data/bball_top50_decimal_removed.csv") 
car_data = pd.read_csv("/Users/emily/Desktop/categorization_experiment-master/data/practice.csv") 
```

```python
def set_type(row):
    if row["Sedan"] == "1":
        return "Sedan"
    elif row["Sports Car"] == "1":
        return "Sports Car"
    elif row["Wagon"] == "1":
        return "Wagon"
    elif row["SUV"] == "1":
        return "SUV"
    else:
        return "Minivan"

car_data = car_data.assign(Type=car_data.apply(set_type, axis=1))

# car_data.head()
```

```python
# calculating accuracy
for user in dict:
    for ans in dict[user]:
        if ans[0][0] == "D":
            dog_id = ans[0][4:]
#             print(dog_id)
            user_label = ans[1]
#             print(user_label)
            true_label = dog_data.loc[dog_data["id"] == int(dog_id)].iloc[0, 1]
#             print(true_label)
            ans.append(true_label)
            if user_label == true_label:
                ans.append("T")
            else:
                ans.append("F")
        if ans[0][0] == "P":
            player_id = ans[0][7:]
#             print(player_id)
            user_label = ans[1]
#             print(user_label)
            true_label = ball_data.loc[ball_data["Player Anonymized"] == int(player_id)].iloc[0, 3]
#             print(true_label)
            ans.append(true_label)
            if user_label == true_label:
                ans.append("T")
            else:
                ans.append("F")
        if ans[0][0] == "C":
            car_id = ans[0][4:]
#             print(car_id)
            user_label = ans[1]
#             print(user_label)
            true_label = car_data.loc[car_data["Car Anonymized"] == int(car_id)].iloc[0, -1]
#             print(true_label)
            ans.append(true_label)
            if user_label == true_label:
                ans.append("T")
            else:
                ans.append("F")
dict
```

```python
# accuracy
for user in dict:
    dog_count, dog_acc = 0, 0 
    ball_count, ball_acc = 0, 0
    car_count, car_acc = 0, 0
    for ans in dict[user]:
        if(ans[0][0]) == 'D':
            dog_count += 1
            if ans[-1] == 'T':
                dog_acc += 1
        if(ans[0][0]) == 'P':
            ball_count += 1
            if ans[-1] == 'T':
                ball_acc += 1
        if(ans[0][0]) == 'C':
            car_count += 1
            if ans[-1] == 'T':
                car_acc += 1
    print("user id:", user)
    print("dog:", dog_acc/dog_count) if dog_count != 0 else print("no dog data")
    print("car:", car_acc/car_count) if car_count != 0 else print("no car data")
    print("basketball:", ball_acc/ball_count) if ball_count != 0 else print("no basketball data")
    print()
```
